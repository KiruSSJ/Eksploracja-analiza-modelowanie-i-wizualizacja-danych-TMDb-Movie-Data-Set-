```python
import requests
from bs4 import BeautifulSoup

# Pobranie danych z strony
url = 'https://inflationdata.com/Inflation/Inflation_Rate/HistoricalInflation.aspx'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Znalezienie tabeli z danymi o inflacji
table = soup.find('table', {'id': 'GridView1'})

# Utworzenie słownika ze średnia inflacją roczną
inflation_dict = {}
for row in table.find_all('tr')[2:]:
    columns = row.find_all('td')
    year = int(columns[0].text)
    rate = float(columns[13].text.replace('%', ''))
    if year not in inflation_dict:
        inflation_dict[year] = []
    inflation_dict[year].append(rate)

```

```python
inflation_dict[2023] = [5.81]
print(inflation_dict)
```

```python
df_inflation = pd.DataFrame(inflation_dict)
df_inflation = df_inflation.sort_index(axis=1)
df_inflation = df_inflation.transpose().reset_index()
df_inflation.columns = ['year', 'rate']

df_inflation.head(10)
```

```python
dfadjusted = df[['id','budget', 'revenue','release_date']]
```

```python
merged_df = pd.merge(dfadjusted, df_inflation, left_on=dfadjusted['release_date'].apply(lambda x: int(str(x)[:4])), right_on='year', how='left')

# Obliczanie wartości na rok 2023
merged_df['adjusted_budget'] = merged_df['budget'] * (merged_df['rate']/100 + 1)**(2023 - merged_df['release_date'].apply(lambda x: int(str(x)[:4])))
merged_df['adjusted_revenue'] = merged_df['revenue'] * (merged_df['rate']/100 + 1)**(2023 - merged_df['release_date'].apply(lambda x: int(str(x)[:4])))

# Wynik
merged_df
```

```python
merged_df['profit'] = merged_df['adjusted_revenue'] - merged_df['adjusted_budget'] 

merged_df
```

Wyciąganie słów kluczowych z kolumny 'overview'.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation


# wczytaj polskie stopwords
stop_words = stopwords.words('polish')

# usun znaki interpunkcyjne z tekstu
translator = str.maketrans('', '', punctuation)
df['clean_overview'] = df['overview'].apply(lambda x: x.translate(translator))

# tokenizuj słowa
df['tokens'] = df['clean_overview'].apply(word_tokenize)

# usun stopwords
df['keywords'] = df['tokens'].apply(lambda x: [word for word in x if word.lower() not in stop_words])

# usun zbedne znaki
df['keywords'] = df['keywords'].apply(lambda x: [word.lower() for word in x if word.isalpha()])

# wyswietl wyniki
print(df['keywords'])

```
